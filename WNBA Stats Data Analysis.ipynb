```python
import pandas as pd
import plotly.express as px
import numpy as np
from nba_api.stats.endpoints import leaguedashplayerstats
import datetime
import time


pd.set_option('display.max_columns', None) #so we can see all columns in a wide DataFrame

# Getting the past 7 years
current_year = datetime.datetime.now().year
all_seasons = []

for i in reversed(range(7)):
    year = current_year - i
    season_str =f'{year-1}-{str(year)[-2:]}'
    stats = leaguedashplayerstats.LeagueDashPlayerStats(
        league_id_nullable = '10',
        season = season_str,
        season_type_all_star = 'Regular Season',
        per_mode_detailed = 'PerGame'
)
    time.sleep(1)

    # Get the data as a Pandas DataFrame
    df = stats.get_data_frames()[0]
    # Data cleaning & analysis preparation
    df = df.rename(columns={'PLAYER_NAME': 'PLAYER', 'TEAM_ABBREVIATION': 'TEAM'})
    df['Year'] = year - 1
    df['Season_type'] = 'Regular Season'
    df = df.drop(columns=['TEAM_ID', 'NICKNAME', 'AGE', 'W', 'L', 'W_PCT', 'BLKA', 'PF', 'PFD', 'PLUS_MINUS', 'NBA_FANTASY_PTS', 'DD2', 'TD3', 'WNBA_FANTASY_PTS', 'GP_RANK', 'W_RANK', 'L_RANK', 'W_PCT_RANK', 'MIN_RANK', 'FGM_RANK', 'FGA_RANK', 'FG_PCT_RANK', 'FG3M_RANK', 'FG3A_RANK', 'FG3_PCT_RANK', 'FTM_RANK', 'FTA_RANK', 'FT_PCT_RANK', 'OREB_RANK', 'DREB_RANK', 'REB_RANK', 'AST_RANK', 'TOV_RANK', 'STL_RANK', 'BLK_RANK', 'BLKA_RANK', 'PF_RANK', 'PFD_RANK', 'PTS_RANK', 'PLUS_MINUS_RANK', 'NBA_FANTASY_PTS_RANK', 'DD2_RANK', 'TD3_RANK', 'WNBA_FANTASY_PTS_RANK', 'TEAM_COUNT'])
    df = df[['Year', 'Season_type', 'PLAYER_ID', 'PLAYER', 'TEAM', 'GP', 'MIN', 'FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PTS']]


    print(f'{season_str}: {df.shape}')
    all_seasons.append(df)


data = pd.concat(all_seasons, ignore_index=False)
data_all = data.copy()
data = data[data['Year'] == current_year - 1]
print(data)
```

```python
total_cols = ['MIN', 'FGM', 'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA',
              'OREB','DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PTS']
```

Which play stats are correlated with each other?

```python
# Aggregate across seasons, then normalize to per-minute rates
data_per_min = (
    data.groupby(['PLAYER', 'PLAYER_ID', 'Year'])[total_cols]
    .sum()
    .reset_index()
)

# Per-minute normalization (skip MIN itself)
for col in total_cols[1:]:
    data_per_min[col] = data_per_min[col] / data_per_min['MIN']

# --- Advanced / efficiency metrics ---
# Guard against division by zero: replace 0 denominators with NaN
fga  = data_per_min['FGA'].replace(0, np.nan)
fg3a = data_per_min['FG3A'].replace(0, np.nan)
fta  = data_per_min['FTA'].replace(0, np.nan)
fgm  = data_per_min['FGM'].replace(0, np.nan)
tov  = data_per_min['TOV'].replace(0, np.nan)

data_per_min['FG%']         = data_per_min['FGM'] / fga
data_per_min['3PT%']        = data_per_min['FG3M'] / fg3a
data_per_min['FT%']         = data_per_min['FTM'] / fta

# FG3A% = share of field-goal attempts that are three pointers (NOT same as 3PT%)
data_per_min['FG3A%']       = data_per_min['FG3M'] / fga
data_per_min['PTS/FGA']     = data_per_min['PTS'] / fga
data_per_min['FG3M/FGM']    = data_per_min['FG3M'] / fgm
data_per_min['FTA/FGA']     = data_per_min['FTA'] / fga

# True Shooting %: measures scoring efficiency accounting for FTs and 3s
data_per_min['TRU%']        = 0.5 * data_per_min['PTS'] / (fga + 0.475 * data_per_min['FTA'])
data_per_min['AST_TOV%']    = data_per_min['AST'] / tov

# Filter to players with meaningful sample (>=10 min per appearance on average)
data_per_min = data_per_min[data_per_min['MIN']>=10].copy()
data_per_min.drop(columns=['PLAYER_ID', 'Year'], inplace=True)

# --- Correlation heatmap ---
corr = data_per_min.corr(numeric_only = True)

fig = px.imshow(
    corr,
    text_auto = '.2f',
    color_continuous_scale = 'RdBu_r',
    zmin = -1, zmax = 1,
    title = 'Per-Minute Stats & Advanced Metrics - Correlation Heatmap',
    width = 900, height = 850,
)
fig.update_layout(coloraxis_colorbar_title = 'r')

fig.show(renderer = 'notebook_connected')
```

How are minutes played distributed?

```python
fig_min = px.histogram(
    data_per_min,
    x = 'MIN',
    nbins = 30,
    title = 'Distribution of Total Played (per player-season)',
    labels = {'MIN': 'Total Minutes'},
    color_discrete_sequence = ['#1d6fa5'],
)
fig_min.add_vline(x = 10, line_dash = 'dash', line_color = 'red', annotation_text = 'MIN Threshold')
fig_min.update_layout()
fig_min.update_layout(bargap = 0.05)
fig_min.show(renderer = 'notebook_connected')
```

How has the game changed over the past 7 years?

```python
# Re-aggregate with Year retained so we can track trends
yearly = (
    data_all[data_all['Season_type'] == 'Regular Season']
    .groupby('Year')[total_cols]
    .mean()
    .reset_index()
)

# Pace/style metrics: 3-point rate, free throw rate, scoring
yearly['3PA_rate'] = yearly['FG3A'] / yearly['FGA'] # share of attempts from the 3
yearly['FTr']      = yearly['FTA'] / yearly['FGA']  # free throw rate

trend_metrics = {
    '3PA_rate': 'Three Point Attempt Rate (FG3A / FGA)',
    'FTr': 'Free Throw Rate (FTA / FGA)',
}
labels = {'value': 'League Average', 'variable': 'Metric'}

yearly_melted = yearly.melt(id_vars='Year', value_vars=['PTS', 'AST', 'REB'])

fig = px.line(yearly_melted, x='Year', y='value', color = 'variable', labels = labels)


for metric, label in trend_metrics.items():
    fig_t = px.line(
        yearly, x = 'Year', y = metric,
        markers = True,
        title = label,
        labels = {'Year': 'Season', metric: label},
    )
    fig_t.update_traces(line_width = 2.5)
    fig_t.show(renderer = 'notebook_connected')

fig.show(renderer = 'notebook_connected')
```
